### 11 ``matplotlib`` 基础

```python
import numpy as np 
x = np.linspace(0, 10, 100) #从0 到10 分100份 
x
```

```python
y = np.sin(x)
y
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(x, y)
```

```python
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="-.")
plt.show()
```

关于linestyle参数：[https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) #x轴范围
plt.ylim(0, 1.5) # y轴范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2]) #对x和y轴同时指定范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)") #图示
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() #加上图示
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!") #标题
plt.show()
```

### Scatter Plot 散点图

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0, 1, 10000) #符合正太分布的数组 均值0 方差1

y = np.random.normal(0, 1, 10000)

plt.scatter(x, y, alpha=0.1) #alpha不透明度 0完全透明 1不透明
plt.show()
```
